```python
# EXERCISE 3: COMPARATIVE ANALYSIS
import numpy as np
import pandas as pd

df = pd.read_csv('programming_languages_popularity.csv', parse_dates=['Month'])

langs = sorted(df.columns[1:])
i = 838 % (len(df.columns)-1)

a, b = langs[i], langs[(i+1) % len(langs)]
df = df[[a, b]]

corr = df[a].corr(df[b])

mean_a, mean_b = df[a].mean(), df[b].mean()
std_a, std_b = df[a].std(), df[b].std()

dom_ratio = (df[a] > df[b]).mean() * 100
rpi = mean_a - mean_b

print(a, "vs", b)
print("Means:", mean_a, mean_b)
print("Std:", std_a, std_b)
print("Correlation:", corr)
print("Dominance %:", dom_ratio)
print("RPI:", rpi)
```

```python
# EXERCISE 2: LIFECYCLE CLASSIFICATION
import numpy as np
import pandas as pd

df = pd.read_csv('programming_languages_popularity.csv', parse_dates=['Month'])

lang = sorted(df.columns[1:])[838 % (len(df.columns)-1)]
s = df[lang]

growth = s.pct_change() * 100
mean_g, std_g = growth.mean(), growth.std()

df = pd.DataFrame({
    'Popularity': s,
    'Growth': growth,
    'MA6': s.rolling(6).mean(),
    'STD6': s.rolling(6).std(),
})

df['Lifecycle'] = np.select(
    [
        (growth > 0) & (growth <= mean_g),
        growth > mean_g,
        growth.abs() <= 1,
        growth < -std_g
    ],
    ['Introduction', 'Growth', 'Maturity', 'Decline'],
    'Maturity'
)

print(df.head())
print(df['Lifecycle'].value_counts())
print("Dominant:", df['Lifecycle'].value_counts().idxmax())
```

```python
# EXERCISE 1: PROGRAMMING LANGUAGE GROWTH ANALYSIS
import numpy as np
import pandas as pd

df = pd.read_csv('programming_languages_popularity.csv', parse_dates=['Month'])

lang = sorted(df.columns[1:])[838 % (len(df.columns)-1)]
s = df[lang]

df = pd.DataFrame({
    'Month': df['Month'],
    'Popularity': s,
    'Growth_Rate': s.pct_change() * 100,
    'MA': s.rolling(3).mean(),
    'STD': s.rolling(3).std(),
})

df['Phase'] = np.select(
    [df['Growth_Rate'] > 5, df['Growth_Rate'] < -5],
    ['Growth', 'Decline'],
    'Stable'
)

print(df.head())
print(df['Popularity'].describe())
print("Growth %:", (s.iloc[-1] - s.iloc[0]) / s.iloc[0] * 100)
print(df['Phase'].value_counts())
```
